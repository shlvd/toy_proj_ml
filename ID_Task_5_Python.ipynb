```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
#load data and create dataframe
data = load_breast_cancer()
train_X = data.data
train_y = data.target
feature_names = data.feature_names
df = pd.DataFrame(data=train_X, columns=feature_names)
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split, cross_val_score, ShuffleSplit
from sklearn.svm import SVC
#create svc object
svc = SVC()
#use first type of cross validation
acc = cross_val_score(svc, train_X, train_y, cv=10)
print('StratifiedKFold')
# print accuracy for each split
for i in range(len(acc)):
    print('Fold' + str(i), 'accuracy ' + str(acc[i]))
print()
# print mean accuracy for all splits
print('Mean accuracy: ',  str(acc.mean()))
print('-'*35)
print('ShuffleSplit')
#use second type of cross validation
shuffle = ShuffleSplit(n_splits=10, test_size=0.1, random_state=0)
acc_sh = cross_val_score(svc, train_X, train_y, cv=shuffle)
# print accuracy for each split
for i in range(len(acc_sh)):
    print('Fold' + str(i), 'accuracy ' + str(acc_sh[i]))
print()
# print mean accuracy for all splits
print('Mean accuracy: ',  str(acc_sh.mean()))
```

Shuffle Split: This function create infinite iterations of data where the test and train are randomly assigned at each iteration. Therefore, we can have a point that is repeated in testing or repeated in training. This could cause issues in ensuring we have a proper validation score if certain classes are over or under represented.

Stratified KFold: This function will shuffle data, then draw boundaries every len(data)/(k+1) observations. We then use one of the k+1 resulting folds as a validation set and train on the rest.

That is, a shuffle split with a 10% test proportion will generate infinitely many randomly split 90/10 train/test buckets. A K=10 fold split will leave you with 11 buckets, of which you treat one as your 10% validation and iterate through 11 times to get a generalized score.

Generally, K folds is seen as the proper method as you prevent any bias from random sampling.
